<h3>Definición de Nodo</h3>

```python
class Nodo:
    def __init__(self, palabra, significado):
        self.palabra = palabra
        self.significado = significado
        self.izq = None
        self.der = None
        self.totalAscii = self.calcular_ascii()   
        self.indice_ascii = 0                   

    def calcular_ascii(self):
        suma = 0
        for caracter in self.palabra:
            suma += ord(caracter)
        return suma

    def sumar_siguiente_ascii(self):
        if self.indice_ascii < len(self.palabra):
            caracter = self.palabra[self.indice_ascii]
            valor = ord(caracter)
            self.totalAscii += valor
            self.indice_ascii += 1
```

<h3>Definición de Árbol Binario</h3>

```python
class ArbolBinario:
    
    # Inicializar el Nodo Raiz en None (vacio)
    def __init__(self):
        self.raiz = None
        
    # Metodo para insertar un nodo en el arbol
    def insertar(self, palabra, significado):
        # Crea un nuevo nodo con la palabra y significado
        nuevo = Nodo(palabra, significado)
        # Si árbol está vacío, nodo se vuelve la raíz.
        if self.raiz is None:
            self.raiz = nuevo
        else:
            # Si no, llama a _insertar_rec() para ubicarlo
            self._insertar_rec(self.raiz, nuevo)
            
    # Función para insertar Nodo en el lugar correcto
    def _insertar_rec(self, actual, nuevo):
        # Mientras tengan el mismo valor ASCII, seguir sumando y comparando
        while nuevo.totalAscii == actual.totalAscii:
            nuevo.sumar_siguiente_ascii()  # Sumar siguiente caracter para desempatar

        # Si son diferentes, decidir hacia dónde va
        # Si la nueva palabra es menor, va al subárbol izquierdo
        if nuevo.totalAscii < actual.totalAscii:
            if actual.izq is None:
                actual.izq = nuevo
            else:
                # Si el lugar ya está ocupado, sigue bajando hasta encontrar
                self._insertar_rec(actual.izq, nuevo)
        else:
            # Si es mayor o igual, va al subárbol derecho.
            if actual.der is None:
                actual.der = nuevo
            else:
                # Si el lugar ya está ocupado, sigue bajando hasta encontrar
                self._insertar_rec(actual.der, nuevo)

                
    # Función recursiva que recorre el árbol en orden (in-order).
    def imprimir_inorder(self):
        self._inorder(self.raiz)

    # Este recorrido visita los nodos en orden ASCII
    def _inorder(self, nodo):
        if nodo:
            # Primero baja todo a la izquierda hasta que no hay nada (nodo = None)
            self._inorder(nodo.izq)
            # Imprime el nodo
            print(f"{nodo.palabra}: {nodo.significado}")
            # Luego sube y va a la derecha
            self._inorder(nodo.der)
            
    # Recorrer e imprimir en ASCII   
    def imprimir_ascii(self):
        self._imprimir_ascii_rec(self.raiz)
    # Este recorrido visita los nodos en orden ASCII
    def _imprimir_ascii_rec(self, nodo):
        if nodo:
            self._imprimir_ascii_rec(nodo.izq)
            print(f"{nodo.palabra} = {nodo.totalAscii}")
            self._imprimir_ascii_rec(nodo.der)
```

<h3>Funcion que extrae la palabra y significado del txt e inicializa el arbol</h3>

```python
# === Cargar palabras desde el TXT e insertar en el árbol ===
def cargar_palabras_en_arbol(ruta_txt):
    arbol = ArbolBinario()
    with open(ruta_txt, "r", encoding="utf-8") as f:
        for linea in f:
            partes = linea.strip().split(":")
            if len(partes) == 2:
                palabra, significado = partes
                arbol.insertar(palabra.strip(), significado.strip())
    return arbol
```

```python
# === Ejecución ===
ruta = "palabras_1000.txt"   # Cambia a 'palabras_1000.txt' si lo deseas
Arbol = cargar_palabras_en_arbol(ruta)
print("=== Palabras en orden ===")
Arbol.imprimir_inorder()
print("=== Palabras en ASCII ===")
Arbol.imprimir_ascii()
```

<h3>Definicion de Clase palabra</h3>

```python
class Palabra:
    texto = ""
    totalAscii = 0
    def __init__(self, texto):
        self.texto = texto
        self.totalAscii = self.calcular_ascii()  # suma total inicial
        self.indice_ascii = 0                    # controla qué letra toca sumar después
        
    # Calcula la suma total de los códigos ASCII de la palabra
    def calcular_ascii(self):
        suma = 0
        for caracter in self.texto:
            suma += ord(caracter)
        return suma
    
    # Suma el código ASCII del siguiente carácter de la palabra a self.totalAscii. Si ya terminó, no hace nada.
    def sumar_siguiente_ascii(self):
        if self.indice_ascii < len(self.texto):
            caracter = self.texto[self.indice_ascii]
            valor = ord(caracter)
            self.totalAscii += valor
            self.indice_ascii += 1
```

<h3>Definicion de Clase Cola</h3>

```python
class Cola:
    def __init__(self):
        self.items = []

    def encolar(self, elemento):
        # Agregar un elemento al final de la cola
        self.items.append(elemento)

    def desencolar(self):
        # Sacar el primer elemento de la cola
        if not self.esta_vacia():
            return self.items.pop(0)

    def esta_vacia(self):
        # Verifica si la cola está vacía
        return len(self.items) == 0
```

```python
def busqueda_anchura_ascii(arbol, palabra_fin):
    if arbol.raiz is None:
        print("El árbol está vacío.")
        return None

    objetivo = Palabra(palabra_fin)  # Inicializar la clase Palabra
    recorrido_general = []           # Guardar todos los recorridos realizados
    intentos = 0                     # Contador de intentos de búsqueda

    while True:
        intentos += 1
        cola = Cola()                # Usamos nuestra clase Cola
        cola.encolar(arbol.raiz)
        recorrido = []

        print(f"\nIntento {intentos} - ASCII objetivo = {objetivo.totalAscii}")

        while not cola.esta_vacia():
            nodo = cola.desencolar()
            recorrido.append(nodo.palabra)

            # Coincidencia en ASCII
            if nodo.totalAscii == objetivo.totalAscii:
                if nodo.palabra.lower() == palabra_fin.lower():
                    print("\nPalabra encontrada")
                    print(f"Palabra: {nodo.palabra}")
                    print(f"Significado: {nodo.significado}")
                    print(f"Total ASCII: {nodo.totalAscii}")
                    print(f"\nRecorrido BFS: {recorrido}")
                    return recorrido

            # Agregar los hijos a la cola si existen
            if nodo.izq:
                cola.encolar(nodo.izq)
            if nodo.der:
                cola.encolar(nodo.der)

        recorrido_general.extend(recorrido)

        # Si no se encontró, intentar ajustar el valor ASCII
        if objetivo.indice_ascii < len(objetivo.texto):
            objetivo.sumar_siguiente_ascii()
            print(f"Ajustando ASCII: nuevo total = {objetivo.totalAscii}")
        else:
            print(f"\nNo se encontró la palabra '{palabra_fin}' tras ajustar todas las letras posibles.")
            print(f"Recorrido completo: {recorrido_general}")
            return None
```

```python
# Buscar una palabra
busqueda_anchura_ascii(Arbol, "Zapatanña")
```

# Búsqueda mejorada en el árbol

Esta implementación realiza una búsqueda eficiente desde una palabra inicial hasta una palabra final en el árbol binario.

```python
def buscar_camino_entre_palabras(arbol, palabra_inicial, palabra_final):
    """
    Busca el camino desde una palabra inicial hasta una palabra final en el árbol.
    Utiliza una búsqueda en anchura (BFS) para encontrar el camino más corto.
    """
    # Primero encontramos el nodo inicial
    if arbol.raiz is None:
        print("El árbol está vacío.")
        return None
        
    # Encontrar el nodo inicial
    nodo_inicial = None
    cola_busqueda = Cola()
    cola_busqueda.encolar(arbol.raiz)
    while not cola_busqueda.esta_vacia():
        nodo = cola_busqueda.desencolar()
        if nodo.palabra.lower() == palabra_inicial.lower():
            nodo_inicial = nodo
            break
        if nodo.izq:
            cola_busqueda.encolar(nodo.izq)
        if nodo.der:
            cola_busqueda.encolar(nodo.der)
    
    if not nodo_inicial:
        print(f"No se encontró la palabra inicial '{palabra_inicial}'")
        return None
    
    # Inicializar estructuras para la búsqueda
    visitados = set()
    cola = Cola()
    cola.encolar(nodo_inicial)
    camino = {}  # Diccionario para reconstruir el camino
    camino[nodo_inicial.palabra] = None  # None indica que es el inicio
    
    print(f"\nComenzando búsqueda desde: {palabra_inicial}")
    print(f"Buscando: {palabra_final}")
    
    # Realizar la búsqueda en anchura
    palabra_encontrada = False
    nodo_final = None
    
    while not cola.esta_vacia():
        nodo_actual = cola.desencolar()
        if nodo_actual.palabra not in visitados:
            visitados.add(nodo_actual.palabra)
            
            print(f"Visitando: {nodo_actual.palabra} (ASCII: {nodo_actual.totalAscii})")
            
            # Verificar si encontramos la palabra objetivo
            if nodo_actual.palabra.lower() == palabra_final.lower():
                palabra_encontrada = True
                nodo_final = nodo_actual
                break
            
            # Agregar nodos hijos a la cola
            for siguiente in [nodo_actual.izq, nodo_actual.der]:
                if siguiente and siguiente.palabra not in visitados:
                    cola.encolar(siguiente)
                    camino[siguiente.palabra] = nodo_actual.palabra
    
    # Si encontramos la palabra, reconstruir y mostrar el camino
    if palabra_encontrada:
        print("\n¡Palabra encontrada!")
        recorrido = []
        actual = nodo_final.palabra
        while actual is not None:
            recorrido.insert(0, actual)
            actual = camino[actual]
        
        print("\nCamino encontrado:")
        print(" -> ".join(recorrido))
        return recorrido
    else:
        print(f"\nNo se encontró un camino desde '{palabra_inicial}' hasta '{palabra_final}'")
        print("Palabras visitadas:", list(visitados))
        return None
```

```python
# Ejemplo de uso
print("=== Buscando camino entre palabras ===")
camino = buscar_camino_entre_palabras(Arbol, "Vorromitan", "Zapatanña")
```
